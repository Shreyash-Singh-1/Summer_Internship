```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from IPython.display import display, HTML


np.random.seed(42)
n_customers = 300

df = pd.DataFrame({
    'CustomerID': range(1, n_customers + 1),
    'Age': np.random.randint(18, 70, n_customers),
    'Annual_Income_k$': np.random.normal(65, 25, n_customers).clip(15, 150),
    'Spending_Score_1_100': np.random.normal(50, 25, n_customers).clip(1, 100)
})


features = ['Age', 'Annual_Income_k$', 'Spending_Score_1_100']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
df['Cluster_ID'] = kmeans.fit_predict(X_scaled)

cluster_names = {
    0: 'Budget Conscious (Low Income, Low Spend)',
    1: 'Target Demographic (High Income, High Spend)',
    2: 'Window Shoppers (High Income, Low Spend)',
    3: 'Impulse Buyers (Low Income, High Spend)'
}
df['Customer Segment'] = df['Cluster_ID'].map(cluster_names)

segment_analysis = df.groupby('Customer Segment')[features].mean().round(1).reset_index()
segment_analysis['Total Customers'] = df.groupby('Customer Segment')['CustomerID'].count().values

display(HTML("<h3>📊 Customer Segment Averages</h3>"))
display(segment_analysis.style.background_gradient(cmap='Blues', subset=['Spending_Score_1_100', 'Annual_Income_k$']))


fig = px.scatter(
    df,
    x='Annual_Income_k$',
    y='Spending_Score_1_100',
    color='Customer Segment',
    size='Age',
    hover_data=['Age'],
    title='🎯 Customer Segmentation Model (2D Bubble Chart)',
    labels={'Annual_Income_k$': 'Annual Income ($k)', 'Spending_Score_1_100': 'Spending Score'},
    opacity=0.7,
    size_max=20,
    color_discrete_sequence=px.colors.qualitative.Bold,
    render_mode='svg'  # <--- This forces standard 2D rendering, bypassing WebGL entirely!
)

fig.update_layout(
    margin=dict(l=20, r=20, b=20, t=40),
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
    template='plotly_white',
    height=600
)

fig.show()
```
