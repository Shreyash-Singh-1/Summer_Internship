```python
import pandas as pd
import numpy as np
import plotly.express as px
from IPython.display import display, HTML
import time

def run_automation_pipeline(data_url):
    print("🚀 Initiating Automated Data Cleaning Pipeline...\n")
    time.sleep(1)

    # 1. EXTRACT (Data Import & Portfolio Trick)
    print("[1/4] Extracting data from GitHub repository...")
    df = pd.read_csv(data_url)

    # Injecting messy data so the pipeline has something to clean for the recruiter!
    messy_duplicates = df.sample(24, random_state=42)
    df = pd.concat([df, messy_duplicates], ignore_index=True)
    df.loc[df.sample(8, random_state=1).index, 'Postal Code'] = np.nan

    initial_rows, initial_cols = df.shape

    # 2. TRANSFORM (Data Cleaning & Formatting)
    print("[2/4] Scrubbing missing values, duplicates, and formatting inconsistencies...")

    if 'Row ID' in df.columns: df = df.drop(columns=['Row ID'])
    df = df.drop_duplicates()
    if 'Postal Code' in df.columns: df['Postal Code'] = df['Postal Code'].fillna('Unknown')

    df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')
    df = df.dropna(subset=['Order Date'])

    text_columns = ['Customer Name', 'City', 'State', 'Category', 'Sub-Category']
    for col in text_columns:
        if col in df.columns: df[col] = df[col].astype(str).str.strip().str.title()

    df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce').fillna(0)

    final_rows, final_cols = df.shape
    rows_removed = initial_rows - final_rows

    # 3. LOAD (Exporting Clean Data)
    print("[3/4] Exporting cleaned dataset...")
    export_filename = 'Cleaned_Superstore_Data.csv'
    df.to_csv(export_filename, index=False)

    # 4. AUTOMATED REPORTING & VISUALIZATION
    print("[4/4] Generating automated execution report...\n")
    time.sleep(1)

    category_summary = df.groupby('Category', as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False)

    report_html = f"""
    <div style="background-color: #f4f6f9; padding: 20px; border-radius: 8px; border-left: 5px solid #0d6efd; color: #000000; font-family: Arial, sans-serif;">
        <h2 style="margin-top: 0; color: #000000;">📑 Automated Data Quality Report</h2>
        <p style="font-size: 15px; color: #000000;"><strong>Execution Status:</strong> <span style="color: #198754;">SUCCESS</span></p>
        <hr style="border: 1px solid #dee2e6;">
        <ul style="font-size: 15px; color: #000000; line-height: 1.6;">
            <li><strong style="color: #000000;">Initial Record Count:</strong> {initial_rows:,} rows</li>
            <li><strong style="color: #000000;">Anomalies/Duplicates Removed:</strong> {rows_removed:,} rows</li>
            <li><strong style="color: #000000;">Final Clean Record Count:</strong> {final_rows:,} rows</li>
            <li><strong style="color: #000000;">Total Revenue Processed:</strong> ${df['Sales'].sum():,.2f}</li>
        </ul>
        <p style="margin-bottom: 0; font-size: 14px; color: #000000;">
            💾 Clean dataset saved as <code>{export_filename}</code>.
        </p>
    </div><br>
    """
    display(HTML(report_html))

    fig = px.bar(category_summary, x='Sales', y='Category', orientation='h', title='📊 Automated Insight: Total Sales by Category', text_auto='.2s', color='Sales', color_continuous_scale='Blues')
    fig.update_layout(template='plotly_white', height=400, yaxis={'categoryorder':'total ascending'})
    fig.show()

github_url = 'https://raw.githubusercontent.com/Shreyash-Singh-1/Summer_Internship/refs/heads/main/train.csv'
run_automation_pipeline(github_url)
```
